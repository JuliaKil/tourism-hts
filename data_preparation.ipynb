# Подготовка данных по кварталам

```python
df = pd.read_excel('data.xls')
```

```python
# отбрасывание столбцов с годовыми показателями
df = df.drop(columns= ['Unnamed: 8', 'Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 23', 'Unnamed: 18', 'Unnamed: 13', 'Unnamed: 28', 'Unnamed: 30'])
```

```python
# создание правильных названий столбцов
df.loc[2,'Unnamed: 4':'Unnamed: 7'] += '2016'
df.loc[2,'Unnamed: 9':'Unnamed: 12'] += '2017'
df.loc[2,'Unnamed: 14':'Unnamed: 17'] += '2018'
df.loc[2,'Unnamed: 19':'Unnamed: 22'] += '2019'
df.loc[2,'Unnamed: 24':'Unnamed: 27'] += '2020'
df.loc[2,'Unnamed: 29'] += '2021'
```

```python
# удаление пустых верхних строк и перемещение строки с названиями столбцов в заголовок
df = df.drop(index=[0,1])
df.columns = df.iloc[0]
df = df[1:]
```

```python
df.columns = df.columns.fillna('Область')
```

```python
# удаление строк, не относящихся к областям
df = df.drop(index=[3, 4, 23, 27, 28, 36, 37, 46, 54, 69, 73, 74, 75, 77, 78, 89, 90, 102])
```

```python
# присвоение областям соответствующих названий округов
df['Округ'] = ''
df.loc[5:22,'Округ'] = 'Центральный федеральный округ'
df.loc[24:35,'Округ'] = 'Северо-Западный федеральный округ'
df.loc[38:45,'Округ'] = 'Южный федеральный округ'
df.loc[47:53,'Округ'] = 'Северо-Кавказский федеральный округ'
df.loc[55:68,'Округ'] = 'Приволжский федеральный округ'
df.loc[70:76,'Округ'] = 'Уральский федеральный округ'
df.loc[79:88,'Округ'] = 'Сибирский федеральный округ'
df.loc[91:101,'Округ'] = 'Дальневосточный федеральный округ'
```

```python
df
```

```python
# превращение кумулятивной суммы по кварталам в пределах года в одельные квартальные значения
df_new = df.copy()
df_new.loc[:,'январь-март2016':'январь-декабрь2016'] = df.loc[:,'январь-март2016':'январь-декабрь2016'].diff(axis=1)
df_new.loc[:,'январь-март2016'] = df.loc[:,'январь-март2016']
df_new.loc[:,'январь-март2017':'январь-декабрь2017'] = df.loc[:,'январь-март2017':'январь-декабрь2017'].diff(axis=1)
df_new.loc[:,'январь-март2017'] = df.loc[:,'январь-март2017']
df_new.loc[:,'январь-март2018':'январь-декабрь2018'] = df.loc[:,'январь-март2018':'январь-декабрь2018'].diff(axis=1)
df_new.loc[:,'январь-март2018'] = df.loc[:,'январь-март2018']
df_new.loc[:,'январь-март2019':'январь-декабрь2019'] = df.loc[:,'январь-март2019':'январь-декабрь2019'].diff(axis=1)
df_new.loc[:,'январь-март2019'] = df.loc[:,'январь-март2019']
df_new.loc[:,'январь-март2020':'январь-декабрь2020'] = df.loc[:,'январь-март2020':'январь-декабрь2020'].diff(axis=1)
df_new.loc[:,'январь-март2020'] = df.loc[:,'январь-март2020']
df_new.loc[:,'январь-март2021'] = df.loc[:,'январь-март2021']
```

```python
#unpivot
df = pd.melt(df_new, id_vars=['Область', 'Округ'])
df = df.rename(columns={2: "Год", "value": "Кол-во"})
```

```python
df
```

```python
df['Год'] = df['Год'].str.replace('январь-март','01.01.')
df['Год'] = df['Год'].str.replace('январь-июнь','01.04.')
df['Год'] = df['Год'].str.replace('январь-сентябрь','01.07.')
df['Год'] = df['Год'].str.replace('январь-декабрь','01.10.')
```
